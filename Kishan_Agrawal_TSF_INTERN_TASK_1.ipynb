***Prediction Using Supervised ML***

```python
#importing libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

### **Reading Data**

```python
data = pd.read_csv("http://bit.ly/w-data")
data.describe()
```

```python
data
```

### **Independent and Dependent Values stored in Variables X , y respectively**

```python
X = data["Hours"].values.reshape(-1,1)
y = data["Scores"].values.reshape(-1,1)
```

### **Visualizing Data**

```python
plt.figure(figsize=(10,8))
plt.scatter(X,y,c="orange")
plt.xlabel("Hours")
plt.ylabel("Scores")
plt.show()
```

### **Splitting Train and Test Data using SK-Learn**

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=101)
```

### **Creating Regression Object**

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
print("Linear Regression Object Created !!")
```

### **Training the Algorithm**

```python
reg.fit(X_train,y_train)
print("Training Complete !!")
```

### **Predicting and Plotting the Regression Line**

```python
y_pred = reg.predict(X_test)
plt.figure(figsize=(10,8))
plt.scatter(X,y,c="grey")
plt.plot(X_test,
        y_pred,
        c="orange",
        linewidth=1.5)
plt.xlabel("Hours")
plt.ylabel("Scores")
plt.show()
```

```python
reg_coef = reg.coef_[0][0]
print("REG COEF : "+str(reg_coef))
reg_intercept = reg.intercept_[0]
print("REG INTERCEPT : "+str(reg_intercept))
```

### **Comparing Actual and Predicted Values**

```python
actual = sum(y_test.tolist(),[])
predicted = sum(y_pred.tolist(),[])
df = pd.DataFrame({' ACTUAL ': actual,' PREDICTED ':predicted})
df
```

### **Calculating R2 Score**

```python
from sklearn.metrics import r2_score
r2_score(y_test,y_pred)
```

## **Accuracy Achieved : 93%**

## **Given Question :**

### **What will be the predicted score if student studies 9.25 hrs/day ?**

```python
hours = float(input("Enter No.of Hours Student Studies : "))
predicted_score = reg.predict([[hours]])
print(f"No of Hours     : {hours} hrs/day")
print(f"Predicted Score : {predicted_score[0][0]}")
```

# Answer :

### **Therefore, If a student Studies for 9.25 hrs/day the predicted score is 92.55**

### **Evaluting Model**

```python
from sklearn import metrics
print("Mean Absolute Error : ", metrics.mean_absolute_error(y_test,y_pred))
print("Max Error : ", metrics.max_error(y_test,y_pred))
print("Mean Squared Error : ", metrics.mean_squared_error(y_test,y_pred))
```
